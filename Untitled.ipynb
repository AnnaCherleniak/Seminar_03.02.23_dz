```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
def f(x):
    return a*x**4*np.sin(np.cos(x)) + b*x**3 + c*x**2 + d*x + e
```

```python
def switch_line():
    global line_style
    if line_style == '-':
        line_style = '-.'
    else:
        line_style = '-'
    return line_style
```

```python
a, b, c, d, e = -12, -18, 5, 10, -30
```

```python
limit = 10
step = 0.01
step_temp = 0.000001
line_style = '-'
color = 'b'
graph_up = True
```

```python
x = np.arange(-limit, limit + step, step)
```

```python
x_points = [(-limit, 'limit')]
for i in range(len(x) - 1):
    if f(x[i]) > 0 and f(x[i + 1]) < 0 or f(x[i]) < 0 and f(x[i + 1]) > 0:
        x_temp = np.arange(x[i], x[i+1] + step_temp, step_temp)
        for j in range(len(x_temp) - 1):
            if f(x_temp[j]) > 0 and f(x_temp[j + 1]) < 0 or f(x_temp[j]) < 0 and f(x_temp[j + 1]) > 0:
                x_points.append((x_temp[j], '0'))
    if graph_up:
        if f(x[i]) > f(x[i+1]):
            x_points.append((x[i], 'down'))
            graph_up = False
    else:
        if f(x[i]) < f(x[i+1]):
            x_points.append((x[i], 'up'))
            graph_up = True
x_points.append((limit, 'limit'))
```

```python
min_y = f(x_points[0][0])
for i in range(len(x_points)):
    if f(x_points[i][0]) < min_y:
        min_y = f(x_points[i][0])
```

```python

for i in range(len(x_points)-1):
    cur_x = np.arange(x_points[i][0], x_points[i+1][0] + step, step)
    if x_points[i][1] == '0':
        plt.plot(x_points[i][0], f(x_points[i][0]), 'gx', label='Корни уравнения')
        plt.rcParams['lines.linestyle'] = switch_line()
        plt.plot(cur_x, f(cur_x), color)
    elif x_points[i][1] == 'down':
        color = 'b'
        plt.plot(cur_x, f(cur_x), color, label='Убывание функции')
    else:
        color = 'r'
        plt.plot(cur_x, f(cur_x), color, label='Возрастание функции')
        if f(x_points[i][0]) == min_y:
            plt.plot(x_points[i][0], min_y, 'yo', label='Вершина на данном участке')

plt.legend()
plt.grid()
```
